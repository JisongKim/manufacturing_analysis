```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset
```

```python
stage1_outputs = df.iloc[:, 50:65].values  # Stage 1 출력값 (15개)
stage2_outputs = df.iloc[:, 100:115].values  # Stage 2 출력값 (15개)

# 입력 데이터 (Machine1~5 센서값)
X_train = df.iloc[:, :50].values

# 데이터 정규화
scaler_X = MinMaxScaler()
scaler_y1 = MinMaxScaler()
scaler_y2 = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X_train)
y_stage1_scaled = scaler_y1.fit_transform(stage1_outputs)
y_stage2_scaled = scaler_y2.fit_transform(stage2_outputs)

# PyTorch Tensor 변환
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_stage1_tensor = torch.tensor(y_stage1_scaled, dtype=torch.float32)
y_stage2_tensor = torch.tensor(y_stage2_scaled, dtype=torch.float32)

# DataLoader 생성
batch_size = 32
dataset1 = TensorDataset(y_stage1_tensor, y_stage1_tensor)  # 🔹 입력을 출력값(15차원)으로 설정
dataset2 = TensorDataset(y_stage2_tensor, y_stage2_tensor)

train_loader1 = DataLoader(dataset1, batch_size=batch_size, shuffle=True)
train_loader2 = DataLoader(dataset2, batch_size=batch_size, shuffle=True)
```

```python
class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 8),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(8, input_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

autoencoder1 = Autoencoder(input_dim=15).to(device)
autoencoder2 = Autoencoder(input_dim=15).to(device)

#loss function과 opotimizer 적용
criterion = nn.MSELoss()
optimizer1 = optim.Adam(autoencoder1.parameters(), lr=0.001)
optimizer2 = optim.Adam(autoencoder2.parameters(), lr=0.001)

y_stage1_tensor = y_stage1_tensor.to(device)
y_stage2_tensor = y_stage2_tensor.to(device)
```

```python
num_epochs = 50

def train_autoenc(model, optimizer, train_loader):
  model.train()
  for epoch in range(num_epochs):
    total_loss = 0
    for data in train_loader:
      inputs, _ = data
      inputs = inputs.to(device)

      optimizer.zero_grad()
      outputs = model(inputs)
      loss = criterion(outputs, inputs)
      loss.backward()
      optimizer.step()
      total_loss += loss.item()

      if (epoch+1) % 10 == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}")
```

```python
train_autoenc(autoencoder1, optimizer1, train_loader1)
train_autoenc(autoencoder2, optimizer2, train_loader2)
```

```python
autoencoder1.eval()
autoencoder2.eval()

# 재구성 오차 계산
with torch.no_grad():
    recon1 = autoencoder1(y_stage1_tensor)
    recon2 = autoencoder2(y_stage2_tensor)

mse1 = torch.mean((y_stage1_tensor - recon1) ** 2, dim=1).cpu().numpy()
mse2 = torch.mean((y_stage2_tensor - recon2) ** 2, dim=1).cpu().numpy()

# 이상 탐지 임계값 설정 (99th percentile)
threshold1 = np.percentile(mse1, 99)
threshold2 = np.percentile(mse2, 99)

# 이상 데이터 판별
df["anomaly_autoencoder_stage1"] = mse1 > threshold1
df["anomaly_autoencoder_stage2"] = mse2 > threshold2

# 이상 데이터 개수 출력
print(df[["anomaly_autoencoder_stage1", "anomaly_autoencoder_stage2"]].sum())
```

```python
# 새로운 라벨 컬럼 추가 (0 = 정상, 1 = 이상)
df["label_stage1"] = df["anomaly_autoencoder_stage1"].astype(int)
df["label_stage2"] = df["anomaly_autoencoder_stage2"].astype(int)

# 정상 데이터 개수 vs 이상 데이터 개수 확인
print("Stage 1 - 정상 데이터 개수:", (df["label_stage1"] == 0).sum())
print("Stage 1 - 이상 데이터 개수:", (df["label_stage1"] == 1).sum())
print("Stage 2 - 정상 데이터 개수:", (df["label_stage2"] == 0).sum())
print("Stage 2 - 이상 데이터 개수:", (df["label_stage2"] == 1).sum())

# 데이터 확인 (일부 샘플 출력)
print(df[["label_stage1", "label_stage2"]].head())
```
